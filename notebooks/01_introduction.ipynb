## Esempio di Dati

Creiamo alcuni dati di esempio per dimostrare le future funzionalitÃ .

````python
# Genera dati temporali di esempio
dates = pd.date_range('2020-01-01', periods=100, freq='D')
values = np.cumsum(np.random.randn(100)) + 100

# Crea DataFrame
df = pd.DataFrame({
    'date': dates,
    'value': values
})

print("Dati di esempio:")
print(df.head())
````

## Visualizzazione

Visualizziamo i dati di esempio.

````python
plt.figure(figsize=(12, 6))
plt.plot(df['date'], df['value'])
plt.title('Serie Temporale di Esempio')
plt.xlabel('Data')
plt.ylabel('Valore')
plt.grid(True)
plt.show()
````
